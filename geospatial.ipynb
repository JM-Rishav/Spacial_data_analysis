```python
def plot_state(statename):
  state_dist = India_Districts[India_Districts.NAME_1 == statename]
  ax=state_dist.plot(cmap='coolwarm', figsize=(10, 10), legend=True)
  plt.show()
```

```python
def plot_district(statename, districtname):
  state_dist = India_Districts[India_Districts.NAME_1 == statename]
  dist = state_dist[state_dist.NAME_2 == districtname]
  ax=dist.plot(cmap='coolwarm',figsize=(10, 10), legend=True)
  plt.show()
```

```python
plot_state(statename='Haryana')
```

```python
plot_district(statename='Haryana', districtname='Gurgaon')
```
